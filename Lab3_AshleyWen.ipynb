```python
import csv
import os

def countries_by_region(input_file):
    try:
        with open(input_file) as csvfile:
            reader = csv.DictReader(csvfile)
            region_files = {}

            for row in reader:
                country_name = row['name'].strip()
                region = row['region'].strip()

                #Local error handling for open/write regional file
                if region not in region_files:
                    try:
                        region_file = open(f"{region}.csv", 'w')
                        writer = csv.writer(region_file)
                        writer.writerow(['name', 'region'])
                        region_files[region] = (writer, region_file)
                    except PermissionError:
                        print("Permission error occurred")
                        continue
                    except IOError as e:
                        print("IO error occurred")
                        continue

                #Write country name and region per each regional file
                region_writer = region_files[region][0]
                region_writer.writerow([country_name, region])

            # Close files
            for value in region_files.values():
              writer, region_file = value
              region_file.close()

        print("Regional files created")

    #Print error messages for different error types
    except FileNotFoundError as fnf_err:
        print(fnf_err)
    except PermissionError as p_err:
        print(p_err)
    except IOError as io_err:
        print(io_err)
    except Exception as e:
        print(e)
```

```python
countries_by_region("country_full.csv")
```
